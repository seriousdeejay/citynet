```python
import pandas as pd
from itertools import combinations
import pandas as pd
import os
import random
import dataframe_image as dfi
```

```python
df = pd.read_csv('cooc_df_with_tops.csv')
```

```python
#limit to top 20 subreddits in terms of co-occurrence comment counts
df2 = pd.DataFrame()
for key in df[' subreddit'].value_counts()[:15].keys():
    df2 = pd.concat([df2, df[df[' subreddit']==key]])

df = df2
cols = [col.lower().replace(' ', '') for col in df.columns]
df.columns = cols
df['subreddit'] = df['subreddit'].apply(lambda x: x.lower())
```

```python
cities = ['paris', 'london', 'madrid', 'berlin', 'milan_merged', 'barcelona',
       'athens_merged', 'roma_merged', 'birmingham', 'lisboa_merged',
       'naples_merged', 'katowice', 'manchester', 'hamburg', 'budapest',
       'bucuresti_merged', 'warsaw_merged', 'stuttgart', 'vienna_merged',
       'munich_merged', 'brussels', 'stockholm', 'frankfurtammain',
       'cologne_merged', 'copenhagen_merged', 'valencia', 'torino_merged',
       'glasgow', 'lyon', 'prague_merged', 'sofia', 'liverpool', 'porto',
       'sevilla_merged', 'dublin', 'helsinki', 'amsterdam', 'rotterdam',
       'd√ºsseldorf']
```

```python
pairs = combinations(cities,2)
pair_list = []
for j in pairs:
    pair = list(j)
    pair.sort()
    pair_list.append(pair)

pair_counts = {}

for i in pair_list:
    pair_string = i[0]+'_'+i[1]
    count_df = df[i]
    count_df['sum'] = count_df.sum(axis = 1)
    count_df = count_df[count_df['sum']==2]
    pair_counts[pair_string] = len(count_df)
```

```python
pair_count_df = pd.DataFrame()
pair_count_df['pair'] = pair_counts.keys()
pair_count_df['count'] = pair_counts.values()
pair_count_df = pair_count_df.sort_values('count', ascending=False)
top_10_pairs = list(pair_count_df[:10]['pair'])
top_10_pairs =[i.replace('_merged', '') for i in top_10_pairs]
```
